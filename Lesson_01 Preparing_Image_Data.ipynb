# Imports and Data

```python
import numpy as np
import PIL
from PIL import Image

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load image
image = Image.open("./images/bondi_beach.jpg")
pixels = np.asarray(image)
```

```python
plt.imshow(pixels)
```

# Original Image Statistics

```python
height = pixels.shape[0]
width = pixels.shape[1]
print("Image height: {}".format(height))
print("Image width:  {}".format(width))
```

```python
# confirm pixel range
datatype = pixels.dtype
print("Data Type:                   {}".format(datatype))
max_pixel = np.max(pixels)
min_pixel = np.min(pixels)
mean_pixel = np.mean(pixels)
std_pixel = np.std(pixels)
print()
print("     Statistic              Value  ")
print("------------------------  ---------")
print("Max Pixel Value:             {}".format(np.max(pixels)))
print("Min Pixel Value:             {}".format(np.min(pixels)))
print("Mean Pixel Value:            {:.2f}".format(np.mean(pixels)))
print("Standard Deviation Value:    {:.2f}".format(np.std(pixels)))
```

uint8: Unsigned Integers (often called "uints") are just like integers (whole numbers) but have the property that they don't have a + or - sign associated with them. Thus they are always non-negative (zero or positive).

```python
# Convert integers to float
pixels = pixels.astype("float32")
datatype = pixels.dtype
print("Data Type:                   {}".format(datatype))
print()
print("     Statistic              Value  ")
print("------------------------  ---------")
print("Max Pixel Value:             {}".format(np.max(pixels)))
print("Min Pixel Value:             {}".format(np.min(pixels)))
print("Mean Pixel Value:            {:.2f}".format(np.mean(pixels)))
print("Standard Deviation Value:    {:.2f}".format(np.std(pixels)))
```

```python
colors = ["r", "g", "b"]
channel_ids= [0,1,2]

plt.xlim(0,256)
for channel_id, color in zip(channel_ids, colors):
    histogram, bin_edges = np.histogram(pixels[:,:, channel_id], bins=256, range=(0,256))
    plt.plot(bin_edges[0:-1], histogram, color=color)

plt.xlabel("Color value", fontsize=14)
plt.ylabel("Pixels",  fontsize=14)

plt.show()
```

# Normalizing Data: Rescaling Images 
Normalize pixel values - rescaling images from the range 0-255 to 0-1

```python
#rescaling
rescaled_pixels = pixels/255.

print("     Statistic              Value  ")
print("------------------------  ---------")
print("Max Pixel Value:             {}".format(np.max(rescaled_pixels)))
print("Min Pixel Value:             {}".format(np.min(rescaled_pixels)))
print("Mean Pixel Value:            {:.2f}".format(np.mean(rescaled_pixels)))
print("Standard Deviation Value:    {:.2f}".format(np.std(rescaled_pixels)))
```

# Standardizing Image

```python
# Standardizing Image
stdized_pixels = (pixels - mean_pixel)/std_pixel
print("         Statistic                      Value  ")
print("-------------------------------------  -------")
print("Mean Intensity of standardized image:   {:.2f}".format(np.mean(stdized_pixels)))
print("STDEV of standardized image:            {:.2f}".format(np.std(stdized_pixels)))
print("Min Intensity of standardized image:    {:.2f}".format(np.min(stdized_pixels)))
print("Max Intensity of standardized image:    {:.2f}".format(np.max(stdized_pixels)))
```

```python
colors = ["r","g","b"]
channel_ids = [0,1,2]

for channel_id, color in zip(channel_ids, colors):
    histogram, bin_edges = np.histogram(stdized_pixels[:,:, channel_id], bins=6, range=(-3,3))
    plt.plot(bin_edges[0:-1], histogram)
plt.xlabel("Color Value", fontsize=14) 
plt.ylabel("Pixels", fontsize=14)
```
